```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('./dc-wikia-data.csv')
```

```python
data.head()
```

```python
data['ALIGN'],_ = data['ALIGN'].factorize()
data['SEX'] = data['SEX'].str.replace(' Characters', '')
```

```python
fig, axes = plt.subplots(2, 2, figsize=(15, 8))
sns.histplot(x='YEAR', data=data, hue = 'SEX', multiple='stack', bins = 50, ax = axes[0,0])
axes[0,0].set_title('Distribution of Gender over Years')
sns.barplot(x = 'SEX', y = 'ALIGN', data = data, ax = axes[0,1])
axes[0,1].set_title('Alignment by Gender')
sns.histplot(x='ALIVE', hue = 'ALIGN', data=data, multiple='stack', ax = axes[1,0])
axes[1,0].set_title('Alive Status by Alignment')
```
