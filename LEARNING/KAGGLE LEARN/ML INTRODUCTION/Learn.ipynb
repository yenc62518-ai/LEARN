```python
import pandas as pd
import numpy as np
data = pd.read_csv('./melb_data.csv')
data
```

```python
data.head()
```

**Chọn Feature và Predict Value**  

Theo quy ước, *Feature* và *Predict Value* được kí hiệu lần lượt là *X* và *y*


```python
y = data.Price
data.corr(numeric_only=True).abs()[["Price"]] 
```

Từ correlation, chọn ra *distance, rooms, bedroom2, bathroom, car, yearbuilt, lattitude, longtitude* để làm features

```python
feature = ['Distance','Rooms', 'Bedroom2', 'Bathroom', 'Car', 'YearBuilt', 'Lattitude', 'Longtitude']
cat = ['Method', 'SellerG', 'CouncilArea']
X = data[feature+cat]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=2)
```

```python
type(x_train)
```

```python
id_train = x_train.index
id_test = x_test.index
```

***IMPUTING VỚI TOÀN BỘ CỘT TRONG DATA***

```python
import sklearn
```

```python
from sklearn.impute import SimpleImputer
x_train[cat] =  sklearn.impute.SimpleImputer(strategy='most_frequent')
x_train[cat] =  sklearn.impute.SimpleImputer(strategy='most_frequent')
SI = SimpleImputer(strategy='mean')
```

```python
SI.fit(x_train[feature])
train = SI.transform(x_train[feature])
test = SI.transform(x_test[feature])
x_train[feature] = pd.DataFrame(train, columns = feature, index = x_train[feature].index)
x_test[feature] = pd.DataFrame(test, columns = feature, index = x_test[feature].index)
```

```python
from sklearn.preprocessing import StandardScaler
SC = StandardScaler()
SC.fit(x_train[feature])
scaled_train = SC.transform(x_train[feature])
scaled_test = SC.transform(x_test[feature])
sc_train = pd.DataFrame(scaled_train, columns=feature, index = id_train)
sc_test = pd.DataFrame(scaled_test, columns=feature, index = id_test)
```

```python
sc_train.head()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
CT = ColumnTransformer(transformers = [('encode', OneHotEncoder(handle_unknown='ignore'), cat)], remainder = 'passthrough')
x_train = CT.fit_transform(x_train)
x_test = CT.transform(x_test)
```

***HÀM CHECK MAX_LEAF***

```python
def check_mae(leaf) :
    from sklearn.ensemble import RandomForestRegressor
    RFR = RandomForestRegressor(
        n_estimators=100, 
        max_depth=10,
        max_leaf_nodes = leaf, 
        random_state=42
    )
    RFR.fit(x_train,y_train)
    raw_pred = RFR.predict(x_test)
    from sklearn.metrics import mean_absolute_error
    x = mean_absolute_error(y_test, raw_pred)
    return x
```

***VALIDATION***

```python
rs = []
```

```python
for i in [5, 50,200, 500, 600 ] :
    rs.append(check_mae(i))
```
